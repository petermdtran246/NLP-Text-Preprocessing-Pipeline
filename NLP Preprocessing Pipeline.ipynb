```python
# -------------------------------------------
# Natural Language Processing (NLP) Preprocessing Pipeline
# Purpose: Clean and prepare raw text for NLP tasks
# -------------------------------------------
```

```python
import nltk
from nltk.tokenize import word_tokenize                       # For advanced tokenization
from nltk.stem import PorterStemmer, WordNetLemmatizer        # For stemming and lemmatization
from nltk.corpus import stopwords                             # Stopwords list
nltk.download('stopwords')                                    # Download stopwords data
import re                                                     # Regular expressions for text cleaning
import pandas as pd                                           # Data handling
```

```python
# 1) Load dataset
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
# Quick overview of dataset structure
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
# 2) Lowercase the text
# Convert full text to lowercase for normalization
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
# 3) Remove stopwords (but keep 'not' for sentiment meaning)
```

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')                 # keep 'not' because it changes sentiment meaning
```

```python
# Remove stopwords using list comprehension + apply
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
data['review_no_stopwords'][0]
```

```python
# 4) Replace star symbol (*) with word "star"
# Some reviews use "4*" instead of "4 star"
```

```python
data['review_no_stopwords_no_punct'] = data.apply(
    lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']), axis=1
)
```

```python
data.head()
```

```python
# 5) Remove remaining punctuation
# Regex: remove any character that is NOT a word character or whitespace
```

```python
data['review_no_stopwords_no_punct'] = data.apply(
    lambda x: re.sub(r"([^\w\s])", "", x['review_no_stopwords_no_punct']), axis = 1
)
```

```python
data.head()
```

```python
# 6) Tokenize text
```

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']),axis=1)
```

```python
data['tokenized'][0]
```

```python
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data['lemmatized'][0]
```

```python
tokens_clean = sum(data['lemmatized'],[])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```
